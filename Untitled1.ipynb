```python
import matplotlib.pyplot as plt
import numpy as np
X = np.array([[1,1],
             [2,2],
              [1.5,2]])
y = np.array([0,1,0])

labels = ["true","false"]
classes = np.unique(y)
colors = plt.cm.coolwarm(np.linspace(0,1,len(classes)))

for cls, color in zip(classes,colors):
    plt.scatter(X[y == cls,0],
                X[y == cls,1],
                color=color,
                label = labels[cls],
                marker = "o")
plt.legend()
plt.show()
```

```python
weights = [1,-1]
bias = 1
function_value = [0,0,0]
datapoints = [0,1,2]
predicted_outputs = [0,0,0]
learning_rate = 0.1
x1_vals = np.linspace(0,3,100)
x2_vals = np.linspace(0,3,100)
X1,X2 = np.meshgrid(x1_vals,x2_vals)

for datapoint_index in datapoints:
    datapoint = X[datapoint_index]
    parameter_0_of_datapoint = datapoint[0]
    parameter_1_of_datapoint = datapoint[1]
    weight_of_parameter_0 = weights[0]
    weight_of_parameter_1 = weights[1]
    function_value[datapoint_index] = parameter_0_of_datapoint*weight_of_parameter_0+parameter_1_of_datapoint*weight_of_parameter_1+bias
    activation_function_value = 1 if function_value[datapoint_index] >= 0 else 0
    predicted_outputs[datapoint_index] = activation_function_value

prediction_error = y - predicted_outputs
bias = bias+learning_rate*sum(prediction_error)
zeroth_parameter_datapoints = X[:,0]
first_parameter_datapoints = X[:,0]

weight_0_delta = sum(learning_rate*prediction_error*zeroth_parameter_datapoints)
weight_1_delta = sum(learning_rate*prediction_error*first_parameter_datapoints)

weights[0] = weights[0]+weight_0_delta
weights[1] = weights[1]+weight_1_delta

w1,w2,b = weights[0],weights[1],bias
Z = w1*X1+w2*X2+b
plt.contourf(X1,X2,Z,levels= [-np.inf,0,np.inf],colors = ['lightblue','lightcoral'],alpha = 0.6)
plt.contour(X1,X2,Z,levels=[0],colors = 'k',linewidths=2)
for cls, color in zip(classes,colors):
    plt.scatter(X[y == cls,0],
                X[y == cls,1],
                color=color,
                label = labels[cls],
                marker = "o")
plt.legend()
plt.show()
```
